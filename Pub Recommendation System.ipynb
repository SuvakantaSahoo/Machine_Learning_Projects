### Pubs by postcode

```python
df_pubs=df[['name','postcode','local_authority','latitude','longitude']]
df_pubs.head(2)
```

```python
desired_pubs = df_pubs[df_pubs['postcode'] == 'CO7 6LW']

desired_pubs
```

```python
# Filter the data to include only the desired pubs
desired_pubs = df_pubs[df_pubs['postcode'] == 'CO7 6LW']

# Create a map centered on the desired area
map_center = [desired_pubs['latitude'].mean(), desired_pubs['longitude'].mean()]

pubs_map = folium.Map(location=map_center, zoom_start=13)

# Add markers for each pub to the map
for index, row in desired_pubs.iterrows():
    pub_location = [row['latitude'], row['longitude']]
    pub_name = row['name']
    pub_marker = folium.Marker(location=pub_location, popup=pub_name)
    pub_marker.add_to(pubs_map)

# Display the map
pubs_map
```

### Finding Nearest Pubs

```python
df_pubs.head(2)
```

```python
# Method 1
user_location = (float(51.970379), float(0.979340))
df_pubs['distance'] = df_pubs.apply(lambda row: ((row['latitude'] - user_location[0])**2 + (row['longitude'] - user_location[1])**2)**0.5, axis=1)
nearest_df = df_pubs.nsmallest(5, 'distance')
nearest_df
```

```python
# Method 2
user_location = (float(51.970379), float(0.979340))
def euclidean_distance(lat1, lon1, lat2, lon2):
    return np.sqrt((lat1-lat2)**2 + (lon1-lon2)**2)

# Calculate the distance between the user's location and each pub in the dataset
df_pubs['distance'] = df_pubs.apply(lambda row: euclidean_distance(row['latitude'], 
                                        row['longitude'],user_location[0],user_location[1]), axis=1)

# Sort the dataset by distance and display the nearest 5 pubs
nearest_pubs = df_pubs.sort_values(by='distance').head(5)
nearest_pubs
```

```python
user_location = (float(51.970379), float(0.979340))

def euclidean_distance(lat1, lon1, lat2, lon2):
    return np.sqrt((lat1-lat2)**2 + (lon1-lon2)**2)

df_pubs['distance'] = df_pubs.apply(lambda row: euclidean_distance(row['latitude'], row['longitude'],
                                                        user_location[0],user_location[1]), axis=1)

nearest_pubs = df_pubs.sort_values(by='distance').head(5)

pubs_map = folium.Map(location=user_location, zoom_start=13)

# Add markers for each pub to the map
for index, row in nearest_pubs.iterrows():
    pub_location = [row['latitude'], row['longitude']]
    pub_name = row['name']
    pub_marker = folium.Marker(location=pub_location, popup=pub_name)
    pub_marker.add_to(pubs_map)

# Display the map
pubs_map
```
